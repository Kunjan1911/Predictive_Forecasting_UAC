```python
# Feature Engineering

## Objectives
#- Create lag features

#- Create rolling statistics
#- Create net pressure indicator
#- Create calendar features
#- Prepare dataset for forecasting models
```

```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
#Load Cleaned Dataset
df = pd.read_csv(
    "../data/cleaned_uac_data.csv"
)
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df.set_index('Date', inplace=True)
df.head()
```

```python
#Creating Lag Features
# HHS care lag 1
df['HHS_Lag1'] = (
    df['Children in HHS Care']
    .shift(1)
)
```

```python
#HHS care lag 2
df['HHS_Lag7'] = (
    df['Children in HHS Care']
    .shift(7)
)
```

```python
#HHS care lag 14
df['HHS_Lag14'] = (
    df['Children in HHS Care']
    .shift(14)
)
```

```python
#Discharge Lag Features
df['Discharge_Lag1'] = (
    df['Children discharged from HHS Care']
    .shift(1)
)

df['Discharge_Lag7'] = (
    df['Children discharged from HHS Care']
    .shift(7)
)

df['Discharge_Lag14'] = (
    df['Children discharged from HHS Care']
    .shift(14)
)
```

```python
#Rolling Meandf['RollingMean7'] = (
df['RollingMean7'] = (
    df['Children in HHS Care']
    .rolling(7)
    .mean()
)
```

```python
df['RollingMean14'] = (
    df['Children in HHS Care']
    .rolling(14)
    .mean()
)
```

```python
#Rolling Standard Deviation
```

```python
df['RollingStd7'] = (
    df['Children in HHS Care']
    .rolling(7)
    .std()
)

df['RollingStd14'] = (
    df['Children in HHS Care']
    .rolling(14)
    .std()
)
```

```python
#Net Pressure Feature
#Net Pressure=Transfers Out of CBP−Discharges from HHS
df['NetPressure'] = (
    df['Children transferred out of CBP custody']
    -
    df['Children discharged from HHS Care']
)
```

```python
#Calendar Features
df['DayOfWeek'] = (
    df.index.dayofweek
)
```

```python
df['Month'] = (
    df.index.month
)
```

```python
df['Quarter'] = (
    df.index.quarter
)
```

```python
df['Weekend'] = (
    df.index.dayofweek >= 5
).astype(int)
```

```python
#Check New Features
df.head(20)
```

```python
#Check Missing Values
df.isnull().sum()
```

```python
df = df.dropna()
print(df.shape)
```

```python
df.to_csv(
    "../data/processed_data.csv"
)
```
